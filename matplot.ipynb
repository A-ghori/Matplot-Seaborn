# Loaded the Data and imports all lib 

```python
import pandas as pd
import matplotlib.pyplot as plt

air_quality = pd.read_csv("/Users/user/Desktop/Developer/jupyter/Matplot-Seaborn/air_quality_no2.csv", index_col=0, parse_dates=True)
```

```python
air_quality
# air_quality.head()


#  air_quality.resample("M")
# 		resample() = Time series data ko new time interval par group karta hai.
# 		"M" ka matlab = Monthly

air_quality.resample("D").mean().plot()
air_quality.shape
```

#  Visually compare the 02 values measured in London versus Paris.

```python

# The alpha parameter controls the transparency level of plot elements like points, lines, bars, etc.
air_quality.plot.scatter(x="station_london", y="station_paris", alpha = 0.5)
```

```python
# Meaning:
# 		dir(air_quality.plot):
# 		Lists all attributes and methods of the air_quality.plot object.
# 		method_name for method_name in ...:
# 		Loops through each item (method/attribute name).
# 		if not method_name.startswith("_"):
# 		Filters out special/private methods like __init__, _private_method, etc.





[
    method_name
    for method_name in dir(air_quality.plot)
    if not method_name.startswith("_") 
]
```

```python
air_quality.plot.box()

# figsize=(12, 4)
# 		Yeh plot ka size define karta hai:
# 		12 units wide
# 		4 units tall
# 		Affects the size of the image that gets rendered.

#  subplots=True
# 		Iska matlab:
# 		Each column of the DataFrame will be plotted separately in its own subplot (one below the other).
# 		Useful when you want to compare multiple measurements but dont want to mix them in one graph.


air_quality.plot.area(figsize=(12,4),subplots=True)
```

```python
#  .mean()
# 		Har month/week/day ke liye average value calculate karta hai.

air_quality.resample("W").mean().plot()
air_quality.resample("M").mean().plot()
```

```python
defi = air_quality.resample("D").std().plot()
defi.set_title("Daily Standard Deviation of NO2 Levels")
```
